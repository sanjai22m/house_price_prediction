```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import load_boston
```

```python
#understanding the dataset
boston=load_boston()
print(boston.DESCR)
```

```python
#access data attributes
dataset=boston.data
for name,index in enumerate(boston.feature_names):
    print(index,name)
```

```python
#reshaping data
data=dataset[:,12].reshape(-1,1)
print(data)

```

```python
np.shape(dataset)
```

```python
target=boston.target.reshape(-1,1)
```

```python
np.shape(target)
```

```python
%matplotlib inline
plt.scatter(data,target,color='blue')
plt.xlabel('lower income population')
plt.ylabel('cost of house')
plt.show()
```

```python
from sklearn.linear_model import Ridge
#creating a regression model
reg=Ridge()
#fitting the model
reg.fit(data, target)
```

```python
pred=reg.predict(data)
```

```python
%matplotlib inline
plt.scatter(data,target,color='red')
plt.plot(data,pred,color='green')
plt.xlabel('lower income population')
plt.ylabel('cost of house')
plt.show()
```

```python
#circumventing curve issue by using polynomial model
from sklearn.preprocessing import PolynomialFeatures

#to allow merging of models (both linear and polynomial)
from sklearn.pipeline import make_pipeline
```

```python
model=make_pipeline(PolynomialFeatures(3),reg)      #PolynomialFeatures(3) is like using a cubic(3) polynomial model for merging
```

```python
model.fit(data,target)
```

```python
pred=model.predict(data)
```

```python
%matplotlib inline
plt.scatter(data,target,color='blue')
plt.plot(data,pred,color='green')
plt.xlabel('lower income population')
plt.ylabel('cost of house')
plt.show()
```

```python
#r2_score metrics
from sklearn.metrics import r2_score
```

```python
r2_score(pred,target)
```

### The r2_score for this prediction model is 0.479
